```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_columns = None
```

```python
nfl = pd.read_csv('~/downloads/pbp-2018.csv')
```

```python
nfl.head()
```

```python
sunday_18 = nfl[nfl['GameDate'] == '2018-09-09'].groupby('GameId')
```

```python
bears_packers = sunday_18.get_group(2018090912)
```

```python
bears_packers.head()
```

```python
plt.scatter('Quarter', 'Yards', data=bears_packers)
```

```python
first = bears_packers[bears_packers['Quarter'] == 1]
first.sort_values(by=['Minute', 'Second'], ascending=False, inplace=True)
first.reset_index(inplace=True)
first.drop(columns = 'index', inplace=True)
```

```python
second = bears_packers[bears_packers['Quarter'] == 2]
second.sort_values(by=['Minute', 'Second'], ascending=False, inplace=True)
second.reset_index(inplace=True)
second.drop(columns = 'index', inplace=True)
```

```python
third = bears_packers[bears_packers['Quarter'] == 3]
third.sort_values(by=['Minute', 'Second'], ascending=False, inplace=True)
third.reset_index(inplace=True)
third.drop(columns = 'index', inplace=True)
```

```python
fourth = bears_packers[bears_packers['Quarter'] == 4]
fourth.sort_values(by=['Minute', 'Second'], ascending=False, inplace=True)
fourth.reset_index(inplace=True)
```

```python
fourth.drop(columns = 'index', inplace=True)
```

```python
fourth[fourth['IsTouchdown'] == 1]
```

```python
full_game = pd.concat([first, second, third, fourth])
```

```python
full_game.reset_index(inplace=True)
```

```python
full_game.drop(columns='index', inplace=True)
```

```python
print(full_game[full_game['IsTouchdown'] == 1]['Description'].values)
```

```python
full_game.groupby(['Quarter', 'OffenseTeam']).sum()['Yards'].plot(kind='bar', figsize=(12,6));
```
