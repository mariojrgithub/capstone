```python
# unseen data
X_final = last_year_df[attributes]
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=False)

X_final_poly = poly.fit_transform(X_final)
```

```python
ss = StandardScaler()

X_final_poly_sc = ss.fit_transform(X_final_poly)
```

Linear Regression Model to Predict AVE

```python
lr_AVE_model.predict(X_final)
```

```python
# accuracy score on new data
lr_test_new = lr_AVE_model.score(X_final, last_year_df['AVE'])
print(lr_test_new)
```

Ridge Model to Predict AVE

```python
ridge_AVE_model.predict(X_final_poly_sc)
```

```python
# accuracy score on new data
ridge_test_new = ridge_AVE_model.score(X_final_poly_sc, last_year_df['AVE'])
print(ridge_test_new)
```

Lasso Model to Predict AVE

```python
lasso_AVE_model.predict(X_final_poly_sc)
```

```python
# accuracy score on new data
lasso_test_new = lasso_AVE_model.score(X_final_poly_sc, last_year_df['AVE'])
print(lasso_test_new)
```

ElasticNet Model to Predict AVE

```python
enet_AVE_model.predict(X_final_poly_sc)
```

```python
# accuracy score on new data
enet_test_new = enet_AVE_model.score(X_final_poly_sc, last_year_df['AVE'])
print(enet_test_new)
```

```python
# function to display R2 scores

def new_data_R2_scores(lr_test_new, ridge_test_new, lasso_test_new, enet_test_new):
#     print(f'Linear Regression RMSE: {lr_score}')
#     print(f'Ridge RMSE: {ridge_score}')
#     print(f'Lasso RMSE: {lasso_score}')
#     print(f'ElasticNet RMSE: {enet_score}')
    print(f'Linear Regression New Data R2 Score: {lr_test_new}')
    
    print(f'Ridge New Data R2 Score: {ridge_test_new}')
   
    print(f'Lasso New Data R2 Score: {lasso_test_new}')
    
    print(f'ElasticNet New Data R2 Score: {enet_test_new}')

    return
    
```

```python
new_data_R2_scores(lr_test_new, ridge_test_new, lasso_test_new, enet_test_new)
```
